```python
%pip install matplotlib qutip
import numpy as np
import matplotlib.pyplot as plt

from qutip import (
    tensor,
    destroy,
    qeye,
)
```

```python
E_C = 0.25
E_J = 20
g = 0.02   # coupling strength (GHz)

N = 6  # truncation per mode

a1 = tensor(destroy(N), qeye(N))
a2 = tensor(qeye(N), destroy(N))

H1 = 4*E_C*(a1.dag()*a1)**2 - E_J*(a1 + a1.dag())
H2 = 4*E_C*(a2.dag()*a2)**2 - E_J*(a2 + a2.dag())

H_int = g * (a1 + a1.dag()) * (a2 + a2.dag())

H = H1 + H2 + H_int

evals = H.eigenenergies()

print("Lowest 5 eigenvalues:")
print(evals[:5])
```

```python
E00 = evals[0]
E01 = evals[1]
E10 = evals[2]
E11 = evals[3]

ZZ = E11 - E10 - E01 + E00
print("Static ZZ (GHz):", ZZ)
```
